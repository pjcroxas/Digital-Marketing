```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt
from datetime import datetime
```

```python
dir = "/content/drive/My Drive/Data Science Portfolio/2024/Digital Marketing/CTR in Advertisement/data"
ctr_df = pd.read_csv(dir+"/Ad_click_prediction_train.csv")
ctr_df.head()
```

```python
print("No. of users: ", ctr_df["user_id"].nunique())
print("No. of products: ", ctr_df["product"].value_counts())
print("No. of product category: ", ctr_df["product_category_1"].nunique())
print("No. of advertisement: ", ctr_df["campaign_id"].nunique())
print("No. of wepbage: ", ctr_df["webpage_id"].nunique())
```

```python
# Convert datetime features
ctr_df["datetime"] = pd.to_datetime(ctr_df["DateTime"])
```

# **Overview of the data**

```python
ad_df = ctr_df.groupby(["product", "campaign_id"]).count()["user_id"].reset_index()
ad_df["campaign_id"] = ad_df["campaign_id"].astype(str)
fig = px.bar(ad_df,
             x="product",
             y="user_id",
             color="campaign_id",
             text_auto=True)

fig.update_layout(template="plotly_white", title="Product in Campaign Ads")
fig.show()
```

- In this viz, we can deduce that `campaign id` is not directly translated to a campaign ad. We can say that a `campaign id` represents who created the campaign, and can be different videos or visuals.
- Among the products offered, product `C` has the highest number of campaigns shows. With over 160k advertisement. This is followed by product `H` and `I` with almost 120k and over 60k respectively.
- The largest number of `campaign_id` in Products `A`, `C`, `F`, `H` is campaign id `405490`.
- The product with the lowest number of campaigns is product `F`, with almost 8k advertisement.

```python
web_df = ctr_df.groupby(["product", "webpage_id"]).count()["user_id"].reset_index()
web_df["webpage_id"] = web_df["webpage_id"].astype(str)
fig = px.bar(web_df,
             x="product",
             y="user_id",
             color="webpage_id",
             text_auto=True)

fig.update_layout(template="plotly_white", title="Product shows in Webpages")
fig.show()
```

- Product `C` has the highest number of presence among websites. Followed by product `H` and `I`.
- `website_id` 13787 has the highest number of ads for Product `C`, `H`, `I`, `B`, and `D`.

### **Successes in advertisement and exposure**

```python
ctr_df.columns
```

```python
ctr_adv = ctr_df.groupby(["product","campaign_id", "is_click"]).count()["user_id"].reset_index()
ctr_adv["campaign_id"] = ctr_adv["campaign_id"].astype(str)
ctr_adv["is_click"] = ctr_adv["is_click"].astype(str)
```

```python
fig = px.bar(ctr_adv,
             x="product",
             y="user_id",
             color="campaign_id",
             text_auto=True,
             facet_row="is_click")

fig.update_layout(template="plotly_white", title="Successes from Campaign Ads", barmode="stack")
fig.show()
```

```python
ctr_web = ctr_df.groupby(["product","webpage_id", "is_click"]).count()["user_id"].reset_index()
ctr_web["webpage_id"] = ctr_web["webpage_id"].astype(str)
ctr_web["is_click"] = ctr_web["is_click"].astype(str)
```

```python
fig = px.bar(ctr_web,
             x="product",
             y="user_id",
             color="webpage_id",
             text_auto=True,
             facet_row="is_click")

fig.update_layout(template="plotly_white", title="Successes from Web Ads", barmode="stack")
fig.show()
```

```python
ctr_date = ctr_df.groupby(['datetime', 'is_click']).count()["user_id"].reset_index()
ctr_date["is_click"] = ctr_date["is_click"].astype(str)
fig = px.line(ctr_date,
              x='datetime',
              y='user_id',
              color="is_click")
fig.update_layout(template="plotly_white")
fig.show()
```
